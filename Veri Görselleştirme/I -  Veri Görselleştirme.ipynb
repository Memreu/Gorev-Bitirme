##  Veri Seti Yükleme ve Ön İzleme

Bu kod parçasında Python kütüphaneleri kullanılarak veri seti yüklenmiş ve incelenmiştir:

- **NumPy (`numpy`)**: Sayısal hesaplamalar için.
- **Pandas (`pandas`)**: Veri işleme ve tablo yapıları için.
- **Matplotlib (`matplotlib`)**: Görselleştirme için.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Jupyter için ayar
%matplotlib inline

# Veri setini yükle
data = pd.read_csv("50_Startups.csv")

# İlk 5 satırı göster
data.head()
```

## Görev 1: R&D harcaması ile kâr arasındaki ilişki  
R&D harcamaları arttıkça kârın nasıl değiştiğini görmek için scatter plot çiziyoruz. Bu ilişki genelde pozitif yönlüdür.  


```python
plt.figure(figsize=(6,4))
plt.scatter(data["R&D Spend"], data["Profit"], color="blue")
plt.xlabel("R&D Harcaması")
plt.ylabel("Kâr")
plt.title("R&D Harcaması ile Kâr Arasındaki İlişki")
plt.show()
```

## Görev 2: Yönetim harcaması ile kâr arasındaki ilişki  
Yönetim (Administration) harcamalarının kâr üzerindeki etkisini scatter plot ile inceliyoruz. Bu harcamanın etkisi R&D kadar güçlü olmayabilir.  



```python
plt.figure(figsize=(6,4))
plt.scatter(data["Administration"], data["Profit"], color="green")
plt.xlabel("Yönetim Harcaması")
plt.ylabel("Kâr")
plt.title("Yönetim Harcaması ile Kâr Arasındaki İlişki")
plt.show()
```

## Görev 3: Eyaletlere göre ortalama kârların karşılaştırılması  
Her eyaletin ortalama kârını hesaplayıp bar chart ile karşılaştırıyoruz. Bu sayede eyalet bazında kârlılık farkları görülebilir.  


```python
state_profit = data.groupby("State")["Profit"].mean()

plt.figure(figsize=(6,4))
state_profit.plot(kind="bar", color=["orange", "purple", "red"])
plt.xlabel("Eyalet")
plt.ylabel("Ortalama Kâr")
plt.title("Eyaletlere Göre Ortalama Kâr")
plt.show()
```

## Görev 4: Harcama kalemlerinin dağılımlarının karşılaştırılması  
R&D, yönetim ve pazarlama harcamalarının dağılımlarını boxplot ile inceliyoruz. Bu grafik, hangi harcama kaleminde uç değerler veya daha geniş bir dağılım olduğunu gösterir.  


```python
plt.figure(figsize=(6,4))
data[["R&D Spend", "Administration", "Marketing Spend"]].plot(kind="box")
plt.title("R&D, Yönetim ve Pazarlama Harcamalarının Dağılımı")
plt.ylabel("Harcama Tutarı")
plt.show()
```

 Genel Yorum:

- 1.Ar-Ge harcamaları düşük ve daha sınırlı bir dağılım gösterirken, pazarlama harcamaları çok daha geniş aralıkta.
- 2.Yönetim harcamaları ise görece sabit kalmış.
- 3.Pazarlama harcamalarının yüksek varyansı, firmaların bu alanda çok farklı stratejiler benimsediğini gösteriyor olabilir.
